```python
"""
Created on Mon Oct 25 09:30:32 2021

@author: Marta Merauje
""""""
This game is based on the Dora The Explorer's espisode called "Baby Winky Comes Home".
In this game, your mission is to help Dora deliver Baby Winky to his family safely.
This game has questions based on general ideas of Dora's routine that happens on
her every mission and critical thinking on helping Baby Winky reach his family before
the spaceship to his planet takes off.
For more detail, viist https://www.youtube.com/watch?v=WHaASkHpYHk
References:
    Kusterer, C. (2021). Jupyter Notebooks-- game starter code. 
"""
##importing necessary packages
from sys import exit
import random

##defining start function
def start():
    #declaring user_name with a new prompt
    #use .capitalize() to control users 
    user_name = input (prompt = 'What is your name?\t').capitalize()
    
    #printing a greeting to user_name
    print (f"\nHi {user_name}, welcome to the Dora The Explorer game!" )
    input(prompt = '<Press Enter to read the instructions>\n')
    
    #printing mission instruction
    print( f"""
{'_' * 40}
MISSION INSTRUCTION:
In this game, you will help Dora and Boots to deliver Baby Winky, a purple baby alien, 
to his family before the spaceship that takes them home to Purple planet takes off.
To win this game, you must be familiar with Dora The Explorer show and have a crtical thinking.        
{'_' * 40}""")
    input('<Now that you know your mission, press Enter to start the game>\n')
    #directing user to the challenge_1 stage
    challenge_1() 

##defining challenge_1 function   
def challenge_1():

#declaring choice_1 with a new prompt
    choice_1 = input(prompt = """
Dora and Boots don't know where Baby Winky's family is.
Who should they ask for help? (Please answer a,b, or c)
A. Swiper
B. Map
C. Backpack
""")
    
    #developing a conditional statement
    if choice_1.casefold() == 'a' or choice_1.casefold == 'swiper':
        #printing a response to the wrong answer
        #adding >>> arows to make responses more visible
        print ("\n>>>Oh no, Swiper can't help. Try again.")
         #directing user to back to challenge_1 stage question
        challenge_1()
        
    elif choice_1.casefold() == 'b' or choice_1.casefold() == 'map':
        #printing a response to the right answer
        print ("\n>>>That's right! The Map can help.\n")
        input(prompt = '<Press Enter to see what the Map has to say>\n')
        
        #printing the Map's explanation
        print(f""" 
{'~' * 80}
Hi there, I'm the Map!

I know where Baby Winky's family is. They are on the top of tallest mountain.
To get to the tallest mountain, they have to go past a farmer's market, 
then hop across hopscotch bridge.
{'~' * 80}
""")
        #directing user to the Map stage
        map() 
    
    elif choice_1.casefold() == 'c' or choice_1.casefold() == 'backpack':
        #printing a response to other wrong answer
        print ("\n>>>Oh no, Backpack can't help. Try again.")
        #directing user to back to challenge_1 stage question
        challenge_1() 
    else:
        #printing an error message if user didnt put acceptable answer
        print("\n>>>Error. Please try again.") 
        #directing user to back to challenge_1 stage question
        challenge_1()
    
    
 #defining map function   
def map():

##declaring choice_2 with a new prompt
    choice_2 = input(prompt ="""\n
<Now you have to help Dora to remember the order to get to the tallest mountain>
  
What is the order to reach the tallest mountain? (Please answer a,b, or c)
A. forest, market, tallest mountain
B. market, lake, tallest mountain
C. market, bridge, tallest mountain
""")
    
    #developing a conditional statement
    if choice_2.casefold() == 'a' or 'forest' in choice_2:
        #printing a response to a wrong answer
        print ("\n>>>Wrong order. Try again.") 
        #prompting user to repeat the question
        map() 
        
    elif choice_2.casefold() == 'b' or 'lake' in choice_2:
        #printing a response to other wrong answer
        print ("\n>>>Not quite. Try again.") 
        map()
    
    elif choice_2.casefold() == 'c' or 'bridge' in choice_2:
        print ("\n>>>That's correct! market, bridge, tallest mountain!")
        print ("\nNow let's go to the market.")
        input(prompt = '<Press Enter to continue>\n')
        #pre market stage prompts
        #put it here so when the user put incorrect answer they dont need to repeat this
        print ("\nHold on, the weather is getting chilly and Baby Winky needs something to keep him warm.")
        print ("\nLet's see if Backpack has something to keep Baby Winky warm.")
        input (prompt = '<Press Enter to see what is in Backpack>\n')
        market()
    else:
        print("\n>>>Invalid command. Please try again.")
        map()
 
 ##defining market function  
def market():
    
    #instantiating a list
    lst = ['Avocado', 'Blanket', 'Camera', 'Drumstick']
    print (lst) 
    
    #declaring objects
    guesses_1 = 3
    
    response_lst = ['\n>>>Try again, you are almost there!.\n',
                   '\n>>>You are so close, try again.\n'
                   '\n>>>Incorrect, try again.\n']
    
    #declaring choice_3 with a new prompt
    choice_3 = input(prompt ='What do they need to keep Baby Winky Warm?\t')
    
    #looping while guesses is greater than zero
    if guesses_1 > 0 and choice_3.casefold() == 'blanket' or 'b' in choice_3:
        #printing results and prompting user to move to the next question
        print ("\n>>>Excellent! You helped Baby Winky to feel warm.")
        input (prompt = '<Press Enter to continue>\n')
        print ("\nThey now have arrived in the market!")
        input (prompt = '<Press Enter to continue>\n')
        print ("\nThey are now on their way to the bridge!")
        input (prompt = '<Press Enter to continue>\n')
        print ("\nThey have arrived at the bridge.")
        bridge()
        
    elif guesses_1 >0 and choice_3.casefold() != 'blanket':
        guesses_1 -=1
        print(random.choice(response_lst)+ f' You have {guesses_1} guesses remaining\n')
        market()
        
    elif guesses_1 == 0:
        print (f"\n>>>You are out of guess. the correct item was blanket")
        fail() 
        #this declaring that user failed
    else:
        print(">>>Error. Try again.")   
        

 ##defining bridge function         
def bridge():
    #printing the bridge stage and its challenge
    print ("""
Bad news! They can't cross the bridge because it has a missing piece.
You have to help Dora and Boots fix the bridge by putting the missing piece in a number order.
The order consists from number 1 (the smallest) to 10 (the biggest). They don't know yet which number is missing. 
Help them figure out by guessing the right number. You have 3 attempts to succeed. 
\n""")   

    missing_number = random.randint (1,10)
    guesses_2 = 3

    while guesses_2 >0:
        #declaring choice_4 with a new prompt
        choice_4 = input (prompt ='Guess from 1 to 10 which number piece is missing\t')

        try:
        #converting choice for numerical comparison
            choice_4 = int(choice_4)
        
        except:
            print("""
>>> Error. Please input a whole number e.g 2 or 6. Try again""")
            bridge()
            
        #conditional statement based on user input
        if choice_4 == missing_number:
            print ('\n>>>Correct! You helped them cross the bridge!')

            mountain()

        elif choice_4 != missing_number:
            print ('\n>>>That is incorrect')
            guesses_2 -=1
            
            #start of nested conditional
            if guesses_2 > 0 and choice_4 < missing_number:
                print(f"\n>>>Number is too low. You have {guesses_2} guessess remaining")

            elif guesses_2 > 0 and choice_4 > missing_number:
                print(f"\n>>>Number is too high. You have {guesses_2} guessess remaining")

            elif guesses_2 == 0:
                print (f"\n>>>You are out of guess. the number was {missing_number}")
                #this declaring that user failed
                fail() 

            else:
                print("\n>>>Error. Try again")
                #end of nested conditional
                
        #setting a different error message for inner and outer loops
        else:
            print("\n>>>Something went wrong.")
            
##defining mountain function
def mountain():   
        print ("\nThey are now on their way to the tallest mountain!")
        input (prompt = "\n<Press Enter 3 times to help them reach the top of the mountain faster>")
        input(prompt = "\n<Press Enter 2 more times!>")
        input(prompt = "\n<Press Enter 1 more time!>")
        print ("""\n
Great job! They are now on top of the tallest mountain and Baby Winky has reunited with his family.

 __     ______  _    _     _____ _____ _____      _____ _______ _ 
 \ \   / / __ \| |  | |   |  __ \_   _|  __ \    |_   _|__   __| |
  \ \_/ / |  | | |  | |   | |  | || | | |  | |     | |    | |  | |
   \   /| |  | | |  | |   | |  | || | | |  | |     | |    | |  | |
    | | | |__| | |__| |   | |__| || |_| |__| |    _| |_   | |  |_|
    |_|  \____/ \____/    |_____/_____|_____/    |_____|  |_|  (_)
                                                                  
                                                                  

""")
    #declaring choice_5 with a new prompt
        choice_5 = input(prompt = 'Press 1 to exit or 2 to replay the game\t')
        if choice_5 == '1' or choice_5.casefold() == 'exit':
            exit(0) 
            #stopping user to play again
            
        else:
            start() 
            #bringing us back to the beginning of start() to try again

##defining fail function  
def fail():
    print("""
 __     __           ______    _ _          _ 
 \ \   / /          |  ____|  (_) |        | |
  \ \_/ /__  _   _  | |__ __ _ _| | ___  __| |
   \   / _ \| | | | |  __/ _` | | |/ _ \/ _` |
    | | (_) | |_| | | | | (_| | | |  __/ (_| |
    |_|\___/ \__,_| |_|  \__,_|_|_|\___|\__,_|
""")                                  
    #declaring choice_6 with a new prompt
    choice_6 = input(prompt = 'Press 1 to exit or 2 to replay the game\t')
    if choice_6 == '1' or choice_6.casefold() == 'exit':
        exit(0)
        #stopping user to play again
        
    else:
        start() 
        #bringing us back to the beginning of start() to try again
        
start()
```
